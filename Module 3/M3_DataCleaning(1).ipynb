```python
#first thing to do is import the following packages

import pandas as pd

#next thing to do is import data set labdata.csv


#when reading file path, make sure to change '\' to '/'.
#backslash is a special character in python, but forwardslash isnt, its safe to switch them here
lab = pd.read_csv("C:/Users/emchege/Desktop/Module 3/labdata.csv")

print(lab)
#we put two spaces in between each print request to make it easier to read
print('\n\n', lab.dtypes)
print('\n\n', lab.count())
```

```python
## Take a look at the data and come up with some data cleaning strategies

#Looking at the data there are a few things to fix
# 1) duplicate rows: subject ID's 103 & 107
# 2) make subject id and income numeric
# 3) splt program_status column int two separate columns: program and status
# 4) restructure treatment and control as values of new column "condition" 
#    and values of these two columns under new column named "score"
# 5) handle missing values
# 6) create dummy variables for categorical data
# 7) standardizng (normalizing) data


```

```python
# 1) duplicated rows: 103 & 107

print(lab.duplicated())
#the duplicated function returns true if the row is a dulicate of another. 
#the drop_duplicates function removes them from the dataframe. it drops the 2nd row that's duplicated, not the first.

lab = lab.drop_duplicates()

print(lab)
```

```python
# 2) make "subject id" and income numeric

#numpy is a python library that deals with large multi-dimensional arrays

import numpy as np

# the to_numeric method will take the replace method as a parameter. this tells it to replace all quotation marks with an empty and treat the string that we're editing as a regular expression
# 1st we remove the quotation marks while making the result a regular expression. 2nd we turn that result into a number. 3rd we save it as lab.subject
lab.subjectid = pd.to_numeric(lab.subjectid.replace('\"','', regex = True))

print(lab)
#we put two spaces in between each print request to make it easier to read
print('\n\n', lab.subjectid.dtypes)
```

```python
# 2) make subject id and "income" numeric

#this is the same as above, but written with more steps to help explain
lab.income = lab.income.replace('\$','', regex = True)

#we only put the '\' before the '$' because '$' is a special character, ',' is not so no '\'
lab.income = lab.income.replace(',','', regex = True)

lab.income = pd.to_numeric(lab.income)

print(lab)
print("\n\n", lab.income.dtypes)
```

```python
# 3) spilt program_status column int two separate columns: program and status

ps_split = lab.program_status.str.split("-")

print(ps_split)

#we split the program status column into two columns, the program and status. 

lab["program"]=ps_split.str.get(0)
lab["status"]=ps_split.str.get(1)

print("\n", lab)

print("\n", lab.dtypes)
```

```python
# 4) restructure treatment and control as values of new column "condition" and values of these two columns under new column named "score"

# we're restructuring two vertical columns into horizontal records to observe if there's a relationship between subjects

#id_vars: the columns that you want to preserve
#value_vars: the columns that you want to change to rows
#value_names: new column you want to store the values of the old columns in
#var_name: how you name the old columns that you turned into rows


lab = pd.melt(frame = lab, id_vars=['subjectid','gender','age','income','program','status','program_status'], value_vars = ['treatment','control'],
              value_name="score",var_name="condition")
#the id_vars here means that we wont be changing: 'subjectid','gender','age','income','program','status','program_status'
#the value_vars means we will be changing the 'treatment' &'control'
#value_name will store the old numbers in the new row: score
#var_name will store the names of the two variables as rows, which correspond to the values in "score". They went from column names to rows.

              
print(lab)
#with treatment scores and control scores in the same columns and labeled properly, we can now compare the two more easily


###Pandas dataframe.melt() function unpivots a DataFrame from wide format to long format, optionally leaving identifier variables set. 
##This function is useful to message a DataFrame into a format where one or more columns are identifier variables (id_vars), while all other columns,
#considered measured variables (value_vars), are “unpivoted” to the row axis, leaving just two non-identifier columns, ‘variable’ and ‘value’.
```

```python
# 5) handle missing values

#we can't compare treatment and control if the values are missing ("the ones that dipsplay NaN")
#so we'll drop the missing ones in the score column with the .dropna function
lab = lab.dropna(subset=['score'])

print(lab.count())
print("\n",lab)
```

```python
# 5) handle missing values

#sometimes we can make a knowledgable guess about missing data. Age is between 19 and 26 so we can enter the mean of all ages for the missing age values

lab = lab.fillna(value={"age":lab.age.mean()})

print(lab.count())
print("\n",lab)
                
```

```python
# 5) handle missing values

#sometimes we're missing too much to make an assumption, we can safely drop the entire row 

lab=lab.dropna(thresh=len(lab.columns)-2)
#measuring how many colums we have, if we're missing enough values in one row, we can drop the row

print(lab.count())
print("\n",lab)
```

```python
#after all of these changes we can clean up the indexes so it doesn't skip numbers

lab = lab.reset_index()

print(lab)
```

```python
# 6) create dummy variables for categorical data

#we have gender listed as M & F, but that isn't numeric data. 
#a lot of algorithms won't take letters, but instead only take numeric values.
#dummy variables are important for handling non-numeric categorical data


dummy_gender=pd.get_dummies(lab['gender']) 
print(dummy_gender.head(3))


#along with gender there are several other non-numeric types of data that we can create dummy variables for

dummy_program=pd.get_dummies(lab['program'])
print(dummy_program.head(3))


dummy_status=pd.get_dummies(lab['status'])
print(dummy_status.head(3))


dummy_condition=pd.get_dummies(lab['condition'])
print(dummy_condition.head(3))


#we're printing the condition with a .head method attached with a 3 as it's perimeter. We just want to see the 1st 3 of each column & it's header, just to see if it works

#It's bad practice to standardize categorical data. Just create dummy variables and standardize the numerical data.
```

```python
# 7) "standardizng" (normalizing) data

#whats the difference between standardized data and normalized data.
#normalized data in alaytics usually means to scale a variable ot have a value between 0 and 1
#standardization usually means to transform the data to have a mean of 0 and a standard deviation of 1. (which is between -1 & +1 with a mean at around 0)
#the number of standard deviations a data point is, away from the mean is called the z-score"

lab_new = lab.drop(columns = ['index', 'subjectid','program_status','gender','condition','program','status']) #we'll 1st make a new datafram with only numercial data

print(lab_new) #the new dataset has only age, income, and score

#Standardizing: panda method ↓

#the formula for standardizing data is:  (numeric value - mean) / std. deviation
lab_norm1 = (lab_new - lab_new.mean()) / lab_new.std()

print(lab_norm1)
#all the data is now standardized to be around either 0, 1, or -1 (some around 2 & -2)
```

```python
#the following formula will also standardize the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

lab_stand2 = pd.DataFrame(scaler.fit_transform(lab_new), index = lab_new.index, columns = lab_new.columns)

print(lab_stand2)
```

```python
# 7) standardizng ("normalizing") data

#Normalizing: panda method ↓

lab_norm1 = (lab_new - lab_new.min()) / (lab_new.max() - lab_new.min())

print(lab_norm1)

#the data is now normalized.
```

```python
#this formula does the same as the formula above
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

lab_norm2 = pd.DataFrame(scaler.fit_transform(lab_new), index = lab_new.index, columns = lab_new.columns)

print(lab_norm2)

```

```python
#Normalize one single column of the dataframe

from sklearn import preprocessing

#we just want to normalize income in this example
income_array = np.array(lab["income"])
#next we create a dataframe to put the normalized data into.

#The .normalize function only takes arrays(it returns an array (or a row))
#the .transpose method switches columns into rows and rows into columns.

normalized_income = pd.DataFrame(preprocessing.normalize([income_array])).transpose()

normalized_income.columns = ["income_normalized"]
lab = lab.join(normalized_income)

print(lab)
#now there's a new column with the normalized income data 
```

```python
#now we'll combine the normalized/standardized data with the dummy variables to make one final clean dataframe
#we appended lab_norm1 just for the example, we could have appended lab_stand1 instead
#axis = 1 means that you're joining them as columns. if axis = 0, it would mean that your appending as rows
lab_final = pd.concat([lab_norm1, dummy_gender], axis = 1) 
lab_final = pd.concat([lab_final, dummy_status], axis = 1)
lab_final = pd.concat([lab_final, dummy_program], axis = 1)
lab_final = pd.concat([lab_final, dummy_condition], axis = 1)

print(lab_final)
```
